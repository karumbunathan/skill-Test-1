```python
#4a
import pandas as pd

data = {
    'Participant':['P1','P2','P3','P4'],
    'Electrode1':[12,15,14,16],
    'Electrode2':[20,18,19,21]
}

df = pd.DataFrame(data)

print("\nOriginal DataFrame:\n")
print(df)

# Mean
mean_values = pd.DataFrame(df.mean(numeric_only=True), columns=['Mean'])
print("\nMean values:\n")
print(mean_values)

# Variance
variance_values = pd.DataFrame(df.var(numeric_only=True), columns=['Variance'])
print("\nVariance values:\n")
print(variance_values)

# Maximum
max_values = pd.DataFrame(df.max(numeric_only=True), columns=['Maximum'])
print("\nMaximum values:\n")
print(max_values)
```
